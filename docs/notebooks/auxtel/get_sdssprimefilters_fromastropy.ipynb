Need pure filter transmission without SDSS telescope transmission because we will apply Auxtel transmission

- Download filters from http://svo2.cab.inta-csic.es/theory/fps/ using astropy tables

```python
from astropy.table import Table 
import matplotlib.pyplot as plt
import numpy as np
```

```python
sdss_u = Table.read("http://svo2.cab.inta-csic.es/theory/fps/fps.php?ID=SLOAN/SDSS.uprime_filter")
sdss_g = Table.read("http://svo2.cab.inta-csic.es/theory/fps/fps.php?ID=SLOAN/SDSS.gprime_filter")
sdss_r = Table.read("http://svo2.cab.inta-csic.es/theory/fps/fps.php?ID=SLOAN/SDSS.rprime_filter")
sdss_i = Table.read("http://svo2.cab.inta-csic.es/theory/fps/fps.php?ID=SLOAN/SDSS.iprime_filter")
sdss_z = Table.read("http://svo2.cab.inta-csic.es/theory/fps/fps.php?ID=SLOAN/SDSS.zprime_filter")
```

```python
df_sdss_u = sdss_u.to_pandas()
df_sdss_g = sdss_g.to_pandas()
df_sdss_r = sdss_r.to_pandas()
df_sdss_i = sdss_i.to_pandas()
df_sdss_z = sdss_z.to_pandas()
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,5))

df_sdss_u.plot(x="Wavelength",y="Transmission",c="b",ax=ax)
df_sdss_g.plot(x="Wavelength",y="Transmission",c="g",ax=ax)
df_sdss_r.plot(x="Wavelength",y="Transmission",c="r",ax=ax)
df_sdss_i.plot(x="Wavelength",y="Transmission",c="orange",ax=ax)
df_sdss_z.plot(x="Wavelength",y="Transmission",c="k",ax=ax)
```

```python
arr_u = np.stack([df_sdss_u["Wavelength"].values/10.,df_sdss_u["Transmission"].values],axis=1)
arr_g = np.stack([df_sdss_g["Wavelength"].values/10.,df_sdss_g["Transmission"].values],axis=1)
arr_r = np.stack([df_sdss_r["Wavelength"].values/10.,df_sdss_r["Transmission"].values],axis=1)
arr_i = np.stack([df_sdss_i["Wavelength"].values/10.,df_sdss_i["Transmission"].values],axis=1)
arr_z = np.stack([df_sdss_z["Wavelength"].values/10.,df_sdss_z["Transmission"].values],axis=1)
```

```python
np.savetxt("SDSS_u_prime.txt",arr_u)
np.savetxt("SDSS_g_prime.txt",arr_g)
np.savetxt("SDSS_r_prime.txt",arr_r)
np.savetxt("SDSS_i_prime.txt",arr_i)
np.savetxt("SDSS_z_prime.txt",arr_z)
```
